```python
class weightedaverage(layers.Layer):
    def __init__(self):
        super(weightedaverage, self).__init__()
    def call(self,inputs):
        res=inputs[0]
        wei=inputs[1]
        weightedsum=tf.math.reduce_sum(tf.multiply(res,wei))
        weightsum=tf.math.reduce_sum(wei)
        return tf.divide(weightedsum,weightsum)
class sumall(layers.Layer):
    def __init__(self):
        super(sumall, self).__init__()
    def call(self,inputs):
        return tf.math.reduce_sum(inputs)
# add constraints

i_pollres = tf.keras.Input(shape=(1,))

i_dbe = tf.keras.Input(shape=(1,))
i_size = tf.keras.Input(shape=(1,))
i_pop = tf.keras.Input(shape=(4,))
i_psterid = tf.keras.Input(shape=(350,))
i_ppoll = tf.keras.Input(shape=(3,))

d_dbe=layers.Dense(1)(i_dbe)
d_size=layers.Dense(1)(i_size)
d_pop=layers.Dense(1,use_bias=False)(i_pop)
d_psterid=layers.Dense(1,use_bias=False)(i_psterid)
d_ppoll=layers.Dense(1,use_bias=False)(i_ppoll)

fweight=layers.multiply([d_dbe,d_size,d_pop,d_psterid,d_ppoll])

wavlayer=weightedaverage()([i_pollres,fweight])

sum1=sumall()(wavlayer)
sum2=sumall()(sum1)

# output=layers.Dot(axes=1)([i_pollres,fweight])

model = tf.keras.Model(inputs=[i_pollres,i_dbe,i_size,i_pop,i_psterid,i_ppoll], outputs=sum2)
model.summary()
```

```python
model.compile(metrics="mean_absolute_error",loss="mean_squared_error",run_eagerly=True)
model.fit(x[0],y[0])
```
